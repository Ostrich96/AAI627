```python
import numpy
dataDir='C:/AAI627/data/'
file_name_test=dataDir + 'testTrack_hierarchy.txt'
file_name_train=dataDir + 'trainIdx2_matrix.txt'
output_file= dataDir + 'output1.txt'
fTest= open(file_name_test, 'r')
fTrain=open(file_name_train, 'r')
Trainline= fTrain.readline()
fOut = open(output_file, 'w')
```

```python
trackID_vec=[0]*6
albumID_vec=[0]*6
artistID_vec=[0]*6
genre1ID_vec = [-1]*6
genre2ID_vec = [-1]*6

genre3ID_vec = [-1]*6
genre4ID_vec = [-1]*6
genre5ID_vec = [-1]*6
genre6ID_vec = [-1]*6
genre7ID_vec = [-1]*6
genre8ID_vec = [-1]*6
genre9ID_vec = [-1]*6
genre10ID_vec = [-1]*6


lastUserID=-1
user_rating_inTrain=numpy.full((6,5),-1)
```

```python
for line in fTest:
    arr_test=line.strip().split('|')
    userID= arr_test[0]
    trackID= arr_test[1]
    albumID= arr_test[2]
    artistID=arr_test[3]
    size = len(arr_test)
    if size>=5:
        genre1ID = arr_test[4]
    if size>=6:
        genre2ID = arr_test[5]
    if size>=7:
        genre3ID = arr_test[6]
    if size>=8:
        genre4ID = arr_test[7]
    if size>=9:
        genre5ID = arr_test[8]
    if size>=10:
        genre6ID = arr_test[9]
    if size>=11:
        genre7ID = arr_test[10]
    if size>=12:
        genre8ID = arr_test[11]
    if size>=13:
        genre9ID = arr_test[12]
    if size>=14:
        genre10ID = arr_test[13]


    if userID!= lastUserID:
        ii=0
        user_rating_inTrain=numpy.full((6,14),-1)
    trackID_vec[ii]=trackID
    albumID_vec[ii]=albumID
    artistID_vec[ii]=artistID
    if 'genre1ID' in locals():
        genre1ID_vec[ii] = genre1ID
    if 'genre2ID' in locals():
        genre2ID_vec[ii] = genre2ID
    if 'genre3ID' in locals():
        genre3ID_vec[ii] = genre3ID  
    if 'genre4ID' in locals():
        genre4ID_vec[ii] = genre4ID 
    if 'genre5ID' in locals():
        genre5ID_vec[ii] = genre5ID 
    if 'genre6ID' in locals():
        genre6ID_vec[ii] = genre6ID 
    if 'genre7ID' in locals():
        genre7ID_vec[ii] = genre7ID 
    if 'genre8ID' in locals():
        genre8ID_vec[ii] = genre8ID 
    if 'genre9ID' in locals():
        genre9ID_vec[ii] = genre9ID 
    if 'genre10ID' in locals():
        genre10ID_vec[ii] = genre10ID  
    
    ii=ii+1
    lastUserID=userID

    if ii==6 : 
        while (Trainline):
        # for Trainline in fTrain:
            arr_train = Trainline.strip().split('|')
            trainUserID=arr_train[0]
            trainItemID=arr_train[1]
            trainRating=arr_train[2]
            Trainline=fTrain.readline()

            if trainUserID< userID:#not find
                continue
            if trainUserID== userID:# found
                for nn in range(0, 6):
                    if trainItemID==albumID_vec[nn]:
                        user_rating_inTrain[nn, 0]=trainRating
                    if trainItemID==artistID_vec[nn]:
                        user_rating_inTrain[nn, 1]=trainRating
                    if trainItemID == genre1ID_vec[nn]:
                        user_rating_inTrain[nn,2] = trainRating
                    if trainItemID == genre2ID_vec[nn]:
                        user_rating_inTrain[nn,3] = trainRating
                    if trainItemID == genre3ID_vec[nn]:
                        user_rating_inTrain[nn,4] = trainRating
                    if trainItemID == genre4ID_vec[nn]:
                        user_rating_inTrain[nn,5] = trainRating
                    if trainItemID == genre5ID_vec[nn]:
                        user_rating_inTrain[nn,6] = trainRating
                    if trainItemID == genre6ID_vec[nn]:
                        user_rating_inTrain[nn,7] = trainRating
                    if trainItemID == genre7ID_vec[nn]:
                        user_rating_inTrain[nn,8] = trainRating
                    if trainItemID == genre8ID_vec[nn]:
                        user_rating_inTrain[nn,9] = trainRating
                    if trainItemID == genre9ID_vec[nn]:
                        user_rating_inTrain[nn,10] = trainRating
                    if trainItemID == genre10ID_vec[nn]:
                        user_rating_inTrain[nn,11] = trainRating
                    
            if trainUserID> userID:
                for nn in range(0, 6):
                    outStr=str(userID) + '|' + str(trackID_vec[nn])+ '|' + str(user_rating_inTrain[nn,0]) + '|' + str(user_rating_inTrain[nn, 1])+ '|' + str(user_rating_inTrain[nn, 2])+ '|' + str(user_rating_inTrain[nn, 3])+ '|' + str(user_rating_inTrain[nn, 4]) +'|'+str(user_rating_inTrain[nn, 5])+ '|'+str(user_rating_inTrain[nn, 6])+ '|'+str(user_rating_inTrain[nn, 7])+ '|'+str(user_rating_inTrain[nn, 8])+ '|'+str(user_rating_inTrain[nn, 9])+ '|'+str(user_rating_inTrain[nn, 10])+ '|'+str(user_rating_inTrain[nn, 11])
                    fOut.write(outStr + '\n')
                break
fTest.close()
fTrain.close()
```
